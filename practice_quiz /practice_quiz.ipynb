- displacement vs. mpg
- weight vs. mpg
- acceleration vs. mpg

````python
# variable 1 vs variable 2 (variable 1 is x and variable 2 is y axis)
# your code here [OKAY]
car_features = pd.read_csv("quiz_data/features.csv")
columns = ["displacement", "weight", "acceleration"]

for feature in columns:
    plt.scatter(x = car_features[feature], y = car_features["mpg"])
    plt.xlabel(feature)
    plt.ylabel("mpg")
   # plt.title(f"Relationship between {feature} and mpg" ) 
    plt.title(feature + " vs mpg")
    plt.show()
````

```python

    coord_y = [2,2, 0, 10, 0, 1]
```

- Create a vector (**vector always means numpy array**) ```hypotenuse_sqr``` with elements $h_k$ defined by the formula

$$ h_k = x_k^2 + y_k^2$$

- You **should not** define it manually. So the following is **not** an acceptable answer:
```python
    hypoteneuse_sqr = np.array([5, 20, 0, 101, 400, 1601])
```

````python
coord_x = [1,4, 0, -1, 20, 40]
coord_y = [2,2, 0, 10, 0, 1]

### Your solution below [OKAY]
x_vector = np.array(coord_x)
y_vector = np.array(coord_y)
hypotenuse_sqr = (x_vector ** 2) + (y_vector **2)
print(hypotenuse_sqr)
````
