Output data for a single column 'cylinders'

<font size="5"> 

Example: Compute a frequency table (a type of "cross-tabulation")

```python
# crosstab counts how many rows fall into categories; rectangular way of presenting data
#frequency tbale is a specific type of cross-tabulation (e.g. male, right on left and top is count (as opposed to example on wiki))
# "index" is the category
# "columns" is a custom title

table = pd.crosstab(index = carfeatures['cylinders'],columns = "count")

display(table)
#only 5 possiblilities of cylinder count on cars
#from table can see, 4,6,8 are more common; 3 and 5 are more rare 
```

### The "help" keyword can be used to learn more about a command

```python
help (pd.crosstab) #not super helpful in the beginning
#needs a first argument for index, a second for columns, can put an argument for values but by default gives none so don't have to put into arguments (= are optional inputs)
```

```python
# It looks like the first two arguments are "index" and "columns", which are required. The rest are optional.
# So we can also write the command like this:

table = pd.crosstab(carfeatures['cylinders'], "count")

display(table)

```

### We can also cross-tabulate between two columns of carfeatures:

```python
table_2 = pd.crosstab(index = carfeatures['cylinders'],columns = carfeatures['mpg'])
display(table_2)
#not a great table because 129 columns for different mpg; a ton of 0s
#would want to aggregate mpg; will learn later on in semester (use ranges like 20-25 mpg e.g.)
```

<font size = "5">

**Q**: How should I change the line of code so that it prints out the 33% and 66% percentile, instead of 25% and 75%?

```python
# your code here 

display(round(carfeatures.describe([0.33,0.66]), 2)) #automatically assigns first argument to percentiles
#this changes the percentile away from the default
#can't get rid of 50% because very important for statistics

display(round(carfeatures.describe(percentiles=[0.33,0.66]), 2))
```

Example: Display a scatter plot 

```python
plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg']) #take carfeatures dataset, grabbing weight column as x input, grabbing mpg input for y input
plt.show() #force it to show plot without extraneous output (at least here)
```

<font size = "5">

**Q**: Is this a good plot?

no because no axis labels or titles

```python
plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg']) 
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship between weight and gas mileage of cars")
plt.show() 
#can say in general, the heavier the car, the worst gas mileage
```

```python
plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg']) 
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship between weight \n and gas mileage of automobiles") #use \n to enter a new line
plt.show()
```

```python
# Try another scatter plot with x = "acceleration"

plt.scatter(x = carfeatures['acceleration'], y = carfeatures['mpg']) 
plt.xlabel("Acceleration")
plt.ylabel("MPG")
plt.title("Relationship between acceleration \n and gas mileage of automobiles") #use \n to enter a new line
plt.show() #relationship here is less clear, weaker relationship

plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg']) 
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relationship between weight \n and gas mileage of automobiles") #use \n to enter a new line
plt.show()

#especially when running multiple plots in same cell, plt.show() very important, or else python will try to cram both plots into one
#plt.show() tells python im completely done with a plot
```
