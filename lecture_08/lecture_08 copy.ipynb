````python
# Now, with a for loop

car_list = ["KIA", "Ferrari", "Ford", "Tesla"] #we are looping over car_list (which is the list); car_name is the loop variable

for car_name in car_list:
    print("Dear customer, we are writing about your " + car_name + " vehicle.")

# The code "for car_name in car_list:", means run through list with each loop variable, running the print code each time

# A variable "car_name" takes the value of every item in the list "car_list", one by one
# For each value of car_name, the indented block of code is executed once.
````

**Example**: Controlling accuracy. Consider the function

$$f(x) = \frac{1}{x}$$

- If $x$ is large enough, then $f(x)$ becomes very small

- We can continue to run values until $0 < f(x) \leq \epsilon$
- Here the value $\epsilon$ is a **threshold**

Similar logic for any algorithm that continues executing until an accuracy criteria is satisfied

````python
#for loops, we know how many times to run
#for while, we don't know but it will keep running until something becomes false

# Specify a threshold level

eps = 0.025

x_list = [] #start with empty lists
y_list = []


x = 1
while 1/x > eps: #once this number gets small enough, we can stop adding numbers to the list (since for this function when x is very large, f(x) approaches infinity), if x is still bigger than my cutoff keep running 
    x_list.append(x) #in debug, after running this list, x_list will have value 1
    y_list.append(1/x)
    x = x + 1

plt.scatter(x = x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
#runs until x=39 because x=40 means 1/x = 0.025 so 1/x > eps becomes false so loop no longer runs anymore
````

## <span style="color:red; font-size:25px"> IV. Interrupting `for` loops </span>

<font size = "4">

Often, instead of a `while` loop, we can use a `for` loop with a `break` statement.

````python
# Specify a threshold level

#using a for loop with a break statement to emulate a while loop

eps = 0.025

x_list = []
y_list = []


x = 1

max_points = 100000 #being conservative here, there is no way going to need more than 100,000 points

for k in range(max_points):
    if 1/x <= eps: #this is the if break statement (if this is true, do next line which is break --> exits the for loop)
        break

    x_list.append(x)
    y_list.append(1/x)
    x = x + 1

plt.scatter(x= x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
eps = 0.25

x_list = []
y_list = []


x = 1

max_points = 100000 #being conservative here, there is no way going to need more than 100,000 points

for k in range(max_points):
    if 1/x <= eps: #this is the if break statement (if this is true, do next line which is break --> exits the for loop, going to line 21) (if false, will skip the indented line and run lines 15, etc.)
        break

    x_list.append(x)
    y_list.append(1/x)
    x = x + 1

plt.scatter(x= x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

<font size = "4">

Example: stop a `for` loop when an invalid type is encountered

````python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2] #the none has no parentheses so an object, not a string

list_sum = 0
for value in list_mixed:
    if (type(value) != int) and (type(value) != float):
        print("Stopping sum computation: Non-numerical element in list!")
        break #synonym for just quit; exit the for loop i am in (even though it belongs in the if block, but because break is within for loop as well -- will go to next line not in the for loop which is line 11)

    list_sum = list_sum + value

print(list_sum)
````

<font size = "4">

In other situations, you might want to silently skip over an invalid input without terminating the entire loop

````python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2, None, "missing", 3]

list_sum = 0 #remember to do this, or it will just carry over from above code's list_sum value!
for value in list_mixed: #loops over list, if it is a number, adds it to the list_sum
    if (type(value) != int) and (type(value) != float):
        print("Warning: Non-numerical element in list!")
        continue # break just means quite what you were doing with loop // continue is skip everything that comes after it but goes back to the top (not just breaking and quitting) -> moves onto the next element in the loop

    list_sum = list_sum + value

print(list_sum)
````

#### **6)**. Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
N = 25
random_array = np.random.uniform(low = -5, high = 10, size = N)

# your code here
sum = 0
#print(random_array)
for element in random_array:
    if element <= 0:
        continue
    sum = sum + element
print(sum)
print("expected solution:", np.sum(random_array[random_array>0]))
````

#### **7)**. In 1734, Leonhard Euler discovered the following identity:

$$ \frac{\pi^2}{6} = \frac{1}{1^2} + \frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \cdots $$

#### The sum on the right is an **infinite** series. Of course we cannot perform infinitely many tasks on a computer.

#### What we **can** do is terminate the summation once the fraction $\dfrac{1}{k^2}$ becomes smaller than a pre-specified tolerance. Compute a truncated approximation to $\dfrac{\pi^2}{6}$ using a `for` loop with a `break` statement

````python
tol = 1e-6 # currently set to 10^(-6). Choose other values

# your code here
max_terms = 1_000_000_000
approximation = 0
for k in range(1, max_terms + 1):
    new_term = 1/(k**2)
    approximation = approximation + new_term
    if new_term < tol:
        break
print(approximation)
````
